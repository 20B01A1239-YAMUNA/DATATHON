```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
df = pd.read_excel('Project Dataset.xlsx')
df
```

```python
len(df)
```

```python
df.head(10)
```

```python
dp= df.iloc[:,1:-1]
y=df[['AccidentOccurance ']]
```

```python
x = pd.get_dummies(dp)
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
p=1/(1+np.e**(0-X_test['Vehicle_Age']))
import matplotlib.pyplot as plt
plt.scatter(X_test['Vehicle_Age'], p, color='red')
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predict = lr.predict(X_test)
predict
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
matrix =pd.DataFrame(confusion_matrix(y_test, predict), columns = ['Predicted No', 'Predicted Yes'], index = ['Actual No', 'Actual Yes'])
```

```python
matrix
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test, predict))
print(accuracy_score(y_test, predict))
```
